#### Problema 1:
Paso 1: Análisis y resolución paso a pasoProblema 3: Resta y resultado. Enunciado: Un número al que se le resta 7 da como resultado 18. Planteamiento: Sea $x$ el número desconocido. Ecuación: "Un número al que se le resta 7" es $x - 7$. "Da como resultado 18" completa la igualdad:$$x - 7 = 18$$Despeje: El 7 está restando, por lo que pasa al otro lado de la igualdad sumando:$$x = 18 + 7$$Solución:$$x = 25$$

```python
# Problema 3: x - 7 = 18
# resultado = 18
# valor_restado = 7

resultado = int(input("Ingrese el resultado: "))
valor_restado = int(input("Ingrese el valor restado: "))

# Para hallar x, sumamos el valor restado al resultado
x = resultado + valor_restado

print(f"El número del Problema 3 es: {x}")
```

#### Problema 4:
Problema 4: Operaciones combinadas (Triple y suma)Enunciado: El triple de un número más cuatro es igual a 31.Planteamiento: Sea $x$ el número. El "triple" significa multiplicar por 3 ($3x$). Ecuación:$$3x + 4 = 31$$ Despeje:Primero movemos el 4 que está sumando: $3x = 31 - 4 \implies 3x = 27$. Luego, el 3 que multiplica a la $x$ pasa dividiendo:$$x = \frac{27}{3}$$Solución:$$x = 9$$

```python
# Problema 4: 3x + 4 = 31
# resultado_final = 31
# suma = 4
# multiplicador = 3

resultado_final = int(input("Ingrese el resultado final: "))
suma = int(input("Ingrese la suma: "))
multiplicador = int(input("Ingrese el multiplicador: "))

# Primero restamos el valor que sumaba, luego dividimos por el triple
x = (resultado_final - suma) / multiplicador

print(f"El número del Problema 4 es: {int(x)}")
```

#### Problema 5:
Problema 5: Dimensiones de un terreno (Perímetro)Enunciado: Un terreno tiene un largo que mide 5 m más que el ancho. Perímetro = 50 m. Planteamiento:Ancho: $w$Largo: $w + 5$Ecuación: El perímetro de un rectángulo es la suma de sus cuatro lados: $2 \times (\text{largo} + \text{ancho})$.$$2((w + 5) + w) = 50$$ Despeje:Simplificamos dentro del paréntesis: $2(2w + 5) = 50$.Dividimos todo entre 2: $2w + 5 = 25$. Restamos 5: $2w = 20$.Dividimos entre 2: $w = 10$. Solución: Ancho: 10 metros. Largo: $10 + 5 =$ 15 metros. Comprobación: $10 + 15 + 10 + 15 = 50$.

```python
# Problema 5: 2(w + (w + 5)) = 50 -> 4w + 10 = 50
# perimetro = 50
# diferencia_largo = 5

perimetro = int(input("Ingrese el perímetro: "))
diferencia_largo = int(input("Ingrese la diferencia entre el largo y el ancho: "))

# Despejando w de la ecuación 4w + (2 * diferencia) = perimetro
# w = (perimetro - 2 * diferencia) / 4
ancho = (perimetro - (2 * diferencia_largo)) / 4
largo = ancho + diferencia_largo

print(f"Las medidas del terreno son: Ancho = {ancho}m, Largo = {largo}m")
```
