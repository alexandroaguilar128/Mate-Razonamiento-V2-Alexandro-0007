#### Problema 1:
Paso 1: Resolución Matemática Paso a Paso1. $x + 8 = 15$ Paso: El 8 está sumando, pasa al otro lado restando.Despeje: $x = 15 - 8 $ Resultado: $x = 7$

```python
# Ecuación 1: x + 8 = 15
num_sumando = int(input("Ingrese el número que se suma (ej. 8): "))
resultado = int(input("Ingrese el resultado de la ecuación (ej. 15): "))
x = resultado - num_sumando
print(f"1. El valor de x es: {x}")
```

#### Problema 3:
3. $3x + 4x = 21$ Paso: Agrupamos términos semejantes (las $x$). $3 + 4 = 7$ .Ecuación simplificada: $7x = 21$Despeje: El 7 multiplica, pasa dividiendo. $x = 21 / 7$ Resultado: $x = 3$

```python
# Ecuación 3: 3x + 4x = 21
coef_x1 = int(input("Ingrese el primer coeficiente de x (ej. 3): "))
coef_x2 = int(input("Ingrese el segundo coeficiente de x (ej. 4): "))
resultado = int(input("Ingrese el resultado de la ecuación (ej. 21): "))

coeficiente_x = coef_x1 + coef_x2
x3 = resultado / coeficiente_x
print(f"3. El valor de x es: {int(x3)}")
```

#### Problema 5:
5. $6a - 2a + a = 20$ Paso: Agrupamos las "a". Recuerda que $+ a$ es lo mismo que $+ 1a$.Operación: $(6 - 2 + 1)a = 5a$ Ecuación simplificada: $5a = 20$Despeje: $a = 20 / 5$ Resultado: $a = 4$

```python
# Ecuación 5: 6a - 2a + a = 20
coef_a1 = int(input("Ingrese el primer coeficiente de 'a' (ej. 6): "))
coef_a2 = int(input("Ingrese el segundo coeficiente de 'a' (ej. -2): "))
coef_a3 = int(input("Ingrese el tercer coeficiente de 'a' (ej. 1): "))
resultado = int(input("Ingrese el resultado de la ecuación (ej. 20): "))

coeficiente_a = coef_a1 + coef_a2 + coef_a3
a5 = resultado / coeficiente_a
print(f"5. El valor de a es: {int(a5)}")
```

#### Problema 7:
7. $4x + 3 = 2x + 11$Paso 1: Pasamos las $x$ a la izquierda. El $2x$ pasa como $-2x$.Paso 2: Pasamos los números a la derecha. El $+3$ pasa como $-3$ .Ecuación agrupada: $4x - 2x = 11 - 3$ Simplificación: $2x = 8$ Resultado: $x = 4$

```python
# Ecuación 7: 4x + 3 = 2x + 11
coef_x_left = int(input("Ingrese el coeficiente de x en el lado izquierdo (ej. 4): "))
const_left = int(input("Ingrese la constante en el lado izquierdo (ej. 3): "))
coef_x_right = int(input("Ingrese el coeficiente de x en el lado derecho (ej. 2): "))
const_right = int(input("Ingrese la constante en el lado derecho (ej. 11): "))

dif_x = coef_x_left - coef_x_right
dif_num = const_right - const_left
x7 = dif_num / dif_x
print(f"7. El valor de x es: {int(x7)}")
```

#### Problema 9:
9. $9m + 4 = 5m + 20$ Paso 1: Agrupamos variables: $9m - 5m = 4m$ .Paso 2: Agrupamos números: $20 - 4 = 16$ .Ecuación simplificada: $4m = 16$ Despeje: $m = 16 / 4$ Resultado: $m = 4$

```python
# Ecuación 9: 9m + 4 = 5m + 20
coef_m_left = int(input("Ingrese el coeficiente de m en el lado izquierdo (ej. 9): "))
const_left_m = int(input("Ingrese la constante en el lado izquierdo (ej. 4): "))
coef_m_right = int(input("Ingrese el coeficiente de m en el lado derecho (ej. 5): "))
const_right_m = int(input("Ingrese la constante en el lado derecho (ej. 20): "))

dif_m = coef_m_left - coef_m_right
dif_num9 = const_right_m - const_left_m
m9 = dif_num9 / dif_m
print(f"9. El valor de m es: {int(m9)}")
```

#### Problema 11:
11 $6a + 5 = 2a + 21$Paso 1: Trasladamos términos: $6a - 2a = 21 - 5$ .Simplificación: $4a = 16$ Despeje: $a = 16 / 4$ Resultado: $a = 4$

```python
# Ecuacion 11: 6a + 5 = 2a + 21
coef_a_left = int(input("Ingrese el coeficiente de 'a' en el lado izquierdo (ej. 6): "))
const_a_left = int(input("Ingrese la constante en el lado izquierdo (ej. 5): "))
coef_a_right = int(input("Ingrese el coeficiente de 'a' en el lado derecho (ej. 2): "))
const_a_right = int(input("Ingrese la constante en el lado derecho (ej. 21): "))

dif_a = coef_a_left - coef_a_right
dif_num11 = const_a_right - const_a_left

a11 = dif_num11 / dif_a
print(f"11. El valor de a es: {int(a11)}")
```
