#### Problema 1:
Paso 1: Resolución Matemática Paso a Paso1. $x + 8 = 15$ Paso: El 8 está sumando, pasa al otro lado restando.Despeje: $x = 15 - 8 $ Resultado: $x = 7$

#### Problema 3:
3. $3x + 4x = 21$ Paso: Agrupamos términos semejantes (las $x$). $3 + 4 = 7$ .Ecuación simplificada: $7x = 21$Despeje: El 7 multiplica, pasa dividiendo. $x = 21 / 7$ Resultado: $x = 3$

```python
# Ecuación 3: 3x + 4x = 21 -> 7x = 21
# coeficiente_x = 3 + 4
# x3 = 21 / coeficiente_x
coeficiente_x = 3 + 4
x3 = 21 / coeficiente_x
print(f"3. El valor de x es: {int(x3)}")
```

#### Problema 7:

#### Problema 9:
